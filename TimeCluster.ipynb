```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df = pd.read_csv("./data/qtdbsel102.txt", header=None, sep='\t')
```

```python
ecg = df.iloc[:, 2].values
ecg = ecg.reshape(len(ecg), -1)
```

# step1 - MinMaxScaler

```python
# standardize
scaler = MinMaxScaler()
std_ecg = scaler.fit_transform(ecg)
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(15,5))
plt.xlabel('time')
plt.ylabel('ECG\'s value')
plt.plot(np.arange(5000), std_ecg[:5000], color='b')
plt.show()
```

# step2 - Sliding Window

```python
w = 60
s = 1
t = 0
limit_r = int((len(ecg)-t-w+1)/s + 1)
```

```python
%%time
period = []
SlideWindowDF = np.empty((0, w))
for r in range(1, limit_r):
    period.append(f'{t+(r-1)*s}~{t+(r-1)*s+w-1}')
    SlideWindowDF = np.vstack([SlideWindowDF, ecg[t+(r-1)*s: t+(r-1)*s+w].T])
```

```python
SlideWindowDF.shape
```

# step3 - Dimensionality Reduction(DR)

```python
def scatter_plot(df):
    plt.scatter(df[:,0], df[:, 1])
    plt.show()
```

## PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
PCA_DR = pca.fit_transform(SlideWindowDF)
```

```python
scatter_plot(PCA_DR)
```

## Kernel-PCA

```python
from sklearn.decomposition import KernelPCA

kpca = KernelPCA(n_components=2)
kPCA_DR = kpca.fit_transform(SlideWindowDF)
```

```python
scatter_plot(kPCA_DR)
```

## t-SNE

```python
from sklearn.manifold import TSNE
```

```python
%%time
tSNE_DR = TSNE(n_components=2).fit_transform(SlideWindowDF)
```

```python
scatter_plot(tSNE_DR)
```

# K-Means

```python
from sklearn.cluster import KMeans
```

```python
nc=5
```

```python
kmeans = KMeans(n_clusters = nc, init = 'k-means++', random_state = 1)
y_kmeans = kmeans.fit_predict(PCA_DR)
```

```python
def label_scatter(df, n, y):
    color = ['red', 'blue', 'yellow', 'green', 'black']
    for i in range(n):
        plt.scatter(df[y == i, 0], df[y == i, 1], c=color[i])
    plt.legend()
    plt.show()
```

```python
label_scatter(PCA_DR, nc, y_kmeans)
```
